## Q3. Make a function to find top bowler(in terms of most wickets) against any team. Take input as team.

```python
def top_bowler(t):
  l=data1['batting_team']==t
  m=data1[l]
  m1=m.dropna(subset=['player_dismissed'])
  m2=m1['dismissal_kind']!='run out'
  m2=m1[m2]
  n=m2.groupby('bowler')
  return n.size().sort_values(ascending=False).head(1)
```

```python
top_bowler('Rising Pune Supergiant')
```

## Q4. Write a function which will take team name as input and return it's record of no. occurance in finals and no. of times won.  [Difficult]

ex.-- if team is CSK-- 
CSK in finals-

Played- 8
Won- 3

If any team which has not played in finals like Gujrat Lions

Gujrat Lions in finals-

Played-0


```python
def occur(team):
  d3=data2.drop_duplicates('season',keep='last')
  m1=d3['team1']==team
  m2=d3['team2']==team
  d3=d3[m1 | m2]
  d4=d3['team1']==team
  d5=d3['team2']==team
  d6=d3[d4 | d5]
  x=d6.shape[0]
  d6=d3['winner']==team
  m3=d3[d6]
  y=m3.shape[0]
  print (team,"in finals - ")
  print ("played =", x)
  print ("Won = ",y)
  return
```

```python
occur('Sunrisers Hyderabad')
```
